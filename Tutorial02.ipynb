```python
import json
import ipywidgets as widgets
from IPython.display import display

# Load students
with open("students.json", "r", encoding="utf-8") as f:
    students = json.load(f)   # list of dicts

# Dropdown with "First Last"
options = [(f"{s['firstName']} {s['lastName']}", i) for i, s in enumerate(students)]
dd = widgets.Dropdown(options=options, description="Student", value=0)

# Where we show details
out = widgets.Output()

# Text box + save button to set favoriteProgram
fav = widgets.Text(description="Favorite")
save_btn = widgets.Button(description="Save")

def render():
    """Show the selected student's data."""
    out.clear_output()
    i = dd.value
    with out:
        print(json.dumps(students[i], indent=2, ensure_ascii=False))

def on_change(change):
    if change["name"] == "value":
        # preload current favorite if exists
        fav.value = students[dd.value].get("favoriteProgram", "")
        render()

dd.observe(on_change, names="value")

def on_save(_):
    i = dd.value
    students[i]["favoriteProgram"] = fav.value
    with open("students.json", "w", encoding="utf-8") as f:
        json.dump(students, f, indent=2, ensure_ascii=False)
    render()

save_btn.on_click(on_save)

# Show UI
display(dd, out, fav, save_btn)
render()
```
