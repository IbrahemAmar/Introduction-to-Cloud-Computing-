```python
import json, pandas as pd
from IPython.display import display
import ipywidgets as widgets

# Load data
with open("students.json","r") as f:
    students = json.load(f)

# Quick table of names + email
df = pd.DataFrame(students)
display(df[["firstName","lastName","Email"]])

# Dropdown of students
options = [(f"{s['firstName']} {s['lastName']}", i) for i, s in enumerate(students)]
dd = widgets.Dropdown(options=options, description="Student:", value=0)

# Output area
out = widgets.Output()

def show_student(i):
    out.clear_output()
    with out:
        # Show all fields of selected student as a vertical table
        s = pd.Series(students[i], name="value").to_frame()
        display(s)

def on_change(change):
    if change["name"] == "value":
        show_student(change["new"])

dd.observe(on_change, names="value")

# Optional: text box to set favoriteProgram and save back to JSON
fav = widgets.Text(description="Favorite:")
save_btn = widgets.Button(description="Save")

def on_save(_):
    i = dd.value
    students[i]["favoriteProgram"] = fav.value
    with open("students.json","w") as f:
        json.dump(students, f, indent=2)
    show_student(i)

save_btn.on_click(on_save)

# Display UI
display(dd, out, fav, save_btn)

# Initial display
show_student(dd.value)
```
